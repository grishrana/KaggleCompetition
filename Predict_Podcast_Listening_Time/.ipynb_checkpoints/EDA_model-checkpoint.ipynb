# Feature Engineering

```python
df_train_test = pd.concat([train,test])
df_train_test
```

```python
df_train_test[df_train_test["Genre"] == "Mystery Matters"]
```

## Label Encoding

```python
lbl_encoder = LabelEncoder()
```

```python
total_mapping = dict()
for col in cat_col:
    mapping = dict()
    lbl_encoder.fit(df_train_test[col])
    df_train_test[col] = lbl_encoder.transform(df_train_test[col])
    for label in lbl_encoder.classes_:
        mapping[str(label)] = lbl_encoder.transform([label])

    print(f"\n{col}\n----\n{mapping}")

    total_mapping.update({col:mapping})
```

```python
df_train_test
```

```python
test = df_train_test.loc[df_train_test["Listening_Time_minutes"].isnull(),"Podcast_Name":"Episode_Sentiment"]
test
```

```python
train = df_train_test.loc[~df_train_test["Listening_Time_minutes"].isnull()]
train
```

# Model Evaluation

```python
X = train.drop(columns=["Listening_Time_minutes"])
y=train["Listening_Time_minutes"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3, random_state=101)
```

```python
def rsme(y_true, y_pred):
    return(np.sqrt(np.mean(np.square(y_true-y_pred))))
```

## Linear Regression

```python
lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
pred_y = lr.predict(X_test)
pred_y
```

```python
rsme(y_test, pred_y)
```

```python
pred_y_test = lr.predict(test)
pred_y_test
```

```python
index=[i for i in range(750000,750000+len(pred_y_test))]
```

```python
pred_lr = pd.DataFrame(data={"id":index,"Listening_Time_minutes":pred_y_test})
pred_lr.set_index("id", inplace=True)
```

```python
pred_lr
```

```python
pred_lr.to_csv("Submission/Submission_lr.csv")
```
